```python
import yfinance as yf
data = yf.download('BTC-USD' , start = "2021-01-01" , end = "2022-05-01")
df = data.copy()
#filter price > 60,000
df[df['Close'] > 60000]

#filter return > 0
df["returns"] = df.Close.pct_change()
df[df.returns > 0]

#create funtion that return > 0 -> print"gain" and < 0 -> print"loss"
def func_ret(data):
    data['returns'] = data.Close.pct_change()
    data.loc[data['returns'] > 0 , 'Gain_and_Loss']  = "Gain"
    data.loc[data['returns'] < 0 , 'Gain_and_Loss'] = "Loss"
    print(f"Number of Gain : {len(data.loc[data['returns'] > 0 ])}")
    print(f"Number of Loss : {len(data.loc[data['returns'] < 0 ])}") #using loc for cutting unqualified condition out
    return data

func_ret(data)
```
