```python
import yfinance as yf
data = yf.download('BTC-USD' , start = "2020-01-01" , end = "2022-01-01")

def sma_crossover(df):
    df['sma_12'] = df.Close.rolling(12).mean()
    df['sma_26'] = df.Close.rolling(26).mean()
    df['trend'] = df['sma_12'] > df['sma_26']
    df.loc[(df['trend'] != df['trend'].shift(1)) & (df['trend'] == True), 'action'] = 'buy'
    df.loc[(df['trend'] != df['trend'].shift(1)) & (df['trend'] == False), 'action'] = 'sell'
    return df

sma_crossover(data)
```

```python
df = sma_crossover(data)
df[df['action'].notnull()]
```
