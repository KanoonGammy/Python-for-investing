```python
#import from YFinance
import yfinance as yf
import datetime
end_date = datetime.datetime.now()
start_date = end_date - datetime.timedelta(729)
data_yf = yf.download('TSLA' , start = start_date , end = end_date , interval = '1h')
```

```python
data_yf
```

```python
#Export Data
#to csv
data_yf.to_csv('TSLA_YF.csv')
from google.colab import files
files.download('TSLA_YF.csv')
```

```python
#Export Data
#to csv
data_yf.to_excel('TSLA_YF.xlsx')
from google.colab import files
files.download('TSLA_YF.xlsx')
```
